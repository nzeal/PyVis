```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import math
import pandas as pd

from statistics import mean


file = open("results18.txt")

plt.figure(figsize=(16,10), dpi= 80)
plt.title("Benchmark STARWALL with Intel Compiler 2018", fontsize=22)
plt.xlabel(r"Number runs", size=20)
plt.ylabel(r"Time spent", size=20)

plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.ylim(15, 30)

with file as f:
    y = np.loadtxt(f, dtype="float", comments="#", skiprows=1, unpack=True)
    y2=[min(y), max(y), mean(y)]
    print(y2)

x2=[0, len(y)]
y2=[mean(y),mean(y)]
p1= plt.plot(y, 'ro')
p2= plt.plot(x2, y2, color='tab:cyan', lw= 2.0)

plt.savefig("/Users/nitshuklayahoo.com/Documents/benchmakr/pic/starwall_2018.jpg")
```
